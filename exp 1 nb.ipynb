SCENARION 1:  E-Commerce Sales Data Analysis
DATA DESCRIPTION

E-commerce sales dataset from Kaggle containing product, customer, and transaction details.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")
```

```python

filepath = "C:\\Users\\Rithesh\Downloads\\data.csv\\data.csv"
df = pd.read_csv(filepath)
# Inspect dataset using head(), tail(), info(), and describe().
print("First Five Rows:\n", df.head(5))
print("\nLast Five Rows:\n", df.tail(5))
print("\nDataFrame Info:")  
print(df.info())
print("\nStatistical Summary:\n", df.describe())
```

```python
# Check for missing values.
print("Missing values from dataset")
print(df.isnull().sum())
```

```python
# Visualize sales per product using bar and line charts.
df["Sales"] = df["Quantity"] * df["UnitPrice"]
sales_per_product = (
    df.groupby("Description")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(sales_per_product.index)
print(sales_per_product.values)
plt.figure(figsize=(20,10))
plt.bar(sales_per_product.index, sales_per_product.values, color='skyblue')
```

Insight:

* A small number of products contribute most of the sales

* Sales distribution is highly skewed

SCENARIO 2: Hospital Patient Records
Examine a patient dataset to identify missing health metrics and visualize distributions.

IN-LAB TASKS:
•	Load the dataset into Pandas.
•	Explore data structure and missing values.
•	Visualize glucose levels and age distribution using histograms and boxplots.
•	Identify patterns in health metrics.

```python
# Load the dataset into Pandas.
fp = r"E:\archive\diabetes.csv"
patientData = pd.read_csv(fp)
# Explore data structure and missing values.
print("Missing values from dataset")
print(patientData.isnull().sum())
```

```python
# Visualize glucose levels and age distribution using histograms and boxplots.
plt.figure(figsize=(12, 5))
plt.title("Glucose Level Distribution")
plt.hist(patientData["Glucose"],bins=20, color='lightgreen', edgecolor='black')
plt.xlabel("Glucose Levels")
plt.ylabel("Frequency")
plt.show()
```

```python
# Age distribution (boxplot)
plt.figure()
plt.boxplot(patientData['Age'])
plt.title('Age Distribution')
plt.ylabel('Age')
plt.show()
```

Insight:

* Higher glucose levels are strongly associated with diabetes

* Older patients show higher diabetes risk

SCENARIO 3: Housing Dataset
Investigate a housing dataset for missing features and relationships between house attributes.

IN-LAB TASKS:
•	Load the dataset and inspect columns.
•	Detect and summarize missing values.
•	Visualize correlations between numerical features using scatter plots and heatmaps.

```python
HouseData = pd.read_csv(r"C:\Users\Rithesh\Downloads\archive (1)\Housing.csv")
# •	Load the dataset and inspect columns.
print("Features of the dataset")
print(HouseData.columns)
# •	Detect and summarize missing values.
print("Missing values from dataset")
print(HouseData.isnull().sum())
```

```python
# Visualize correlations between numerical features using scatter plots and heatmaps.
plt.figure()
sns.scatterplot(x=HouseData["area"], y=HouseData["price"],hue=HouseData["bedrooms"])
plt.show()
plt.figure()
sns.scatterplot(x=HouseData["area"], y=HouseData["bedrooms"],hue=HouseData["bathrooms"])
plt.show()
```

SCENARIO 4: Banking Customer Data
Analyze banking data to detect missing account information and understand customer demographics.

IN-LAB TASKS:
•	Import dataset into Pandas.
•	Examine structure, null values, and column types.
•	Visualize age, income, and spending patterns using bar and box plots.
•	Summarize insights from visualizations.

```python
# Import dataset into Pandas with correct separator
BankData = pd.read_csv(r"E:\marketing_campaign.csv", sep='\t')
# •	Examine structure, null values, and column types.
print("Dataset Information:")
BankData.info()
plt.figure()
BankData['Education'].value_counts().plot(kind='bar')
plt.title("Education Distribution of Customers")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.show()
plt.figure()
plt.title("Marital status distribution")
plt.bar(BankData['Marital_Status'].value_counts().index, BankData['Marital_Status'].value_counts().values)
plt.show()
plt.figure()
plt.bar(BankData["Education"].value_counts().index, BankData["Education"].value_counts().values)
plt.show()
plt.figure()
plt.hist(BankData["Year_Birth"], bins=20)
plt.xlabel("Year of Birth")
plt.ylabel("Number of Customers")
plt.title("Year of Birth Distribution of Bank Customers")
plt.show()
```

```python
# 2. Income Distribution (Box Plot)

plt.figure()
sns.boxplot(y=BankData["Income"])
plt.ylabel("Income")
plt.title("Income Distribution")
plt.show()
```

```python
# 3. Spending / Interaction Pattern (Recency Box Plot)

plt.figure()
sns.boxplot(y=BankData["Recency"])
plt.ylabel("Recency (days since last purchase)")
plt.title("Customer Recency Distribution")
plt.show()
```

insights

* Most customers fall into middle-age groups

* Income distribution contains outliers

* Wine and meat products dominate spending

* Missing income values should be handled before modeling
